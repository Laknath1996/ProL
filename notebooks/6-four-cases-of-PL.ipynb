```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
font = {'size':16}
matplotlib.rc('font', **font)
matplotlib.rcParams['figure.facecolor'] = 'white'
import seaborn as sns
import math
import numpy as np
```

```python
def binomial_cdf(n, p, x):
    k_list = np.arange(0, np.floor(x)+1, 1).astype('int')
    prob = 0
    for k in k_list:
        prob += math.comb(n, k) * p**k * (1-p)**(n-k)
    return prob
```

```python
p = 0.8
t_list = np.arange(2, 50, 2)
expected_risk = [(1-p)*(1 - binomial_cdf(t, p, 0.5*t)) + p*binomial_cdf(t, p, 0.5*t) for t in t_list]
```

```python
plt.plot(t_list, expected_risk)
plt.xlabel('t')
plt.ylabel(r'Expected Average Future Risk' + '\n' + r'$\mathbb{E}_{Z_{1:t}}[R_t(h_{\geq t})]$')
plt.title(f"p = {p}")
```
