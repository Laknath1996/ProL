```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
import matplotlib
font = {'size':16}
matplotlib.rc('font', **font)
matplotlib.rcParams['figure.facecolor'] = 'white'
import seaborn as sns
from scipy.linalg import expm
```

```python
A = np.array([[-5, 1], [4, -2]])
```

```python
for _ in range(20):
    x_0 = np.array([2, 1])
    t_list = np.arange(0, 10, 0.1)
    x_list = []
    for t in t_list:
        x = expm(A*t) @ x_0 + np.random.normal(0, scale=0.1, size=(2))
        x_list.append(x)
    x_list = np.array(x_list)
    plt.plot(t_list, x_list[:, 1])
```

```python
from scipy.signal import periodogram
```

```python
x = x_list[:, 1]
```

```python
f, Pxx_den = periodogram(x, fs=1)
```

```python
plt.plot(f, Pxx_den)
plt.xlabel("frequency (Hz)")
plt.ylabel("PSD")
```
