## Time-encoding matrices

```python
from prol.models import proformer
```

```python
model_kwargs = proformer.model_defaults('mnist')
model_kwargs['encoding_type'] = 'freq'
freq_model = proformer.Model(
            num_classes=2,
            **model_kwargs
        )

model_kwargs = proformer.model_defaults('mnist')
model_kwargs['encoding_type'] = 'vanilla'
vanilla_model = proformer.Model(
            num_classes=2,
            **model_kwargs
        )
```

```python
freq = freq_model.pe.detach().cpu().numpy().squeeze()
vanilla = vanilla_model.pe.detach().cpu().numpy().squeeze()

fig, ax = plt.subplots(1, 2, figsize=(20, 10))

fig.suptitle("Time-embedding matrices", fontweight='bold', fontsize=17)

ax[0].matshow(vanilla[:400].T)
ax[0].set_title("Original Transformer", fontsize=15)
ax[0].set_ylabel("Dimension", fontsize=14)
ax[0].set_xlabel("Time", fontsize=14)
ax[1].matshow(freq[:400].T)
ax[1].set_title("Ours", fontsize=15)
ax[1].set_ylabel("Dimension", fontsize=14)
ax[1].set_xlabel("Time", fontsize=14)

for a in ax:
    a.tick_params(labelsize=13)

plt.subplots_adjust(wspace=0.14, top=1.45)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

# Synthetic tasks
files = [
    'experiments/proformer/multirun/2024-05-21/06-09-30',
    'experiments/proformer/multirun/2024-05-21/06-09-48',
    'experiments/proformer/multirun/2024-05-21/06-09-34',
    'experiments/proformer/multirun/2024-05-21/06-09-59'
]

names = [
    'Prosp. Transformer (freq.)',
    'Autoreg. Transformer (freq.)',
    'Prosp. Transformer (vanilla)',
    'Autoreg. Transformer (vanilla)'
]

colors = [
    sns.color_palette("Paired")[1],
    sns.color_palette("Paired")[0],
    sns.color_palette("Paired")[5],
    sns.color_palette("Paired")[4],
]

N = 10

k = 0
for file, name in zip(files, names):
    if file:
        time_list = []
        risk_list = []
        ci_list = []
        for i in range(10):
            fname = f'{file}/{i}/outputs.pkl'
            with open(fname, 'rb') as f:
                outputs = pickle.load(f)
            time_list.append(outputs['args']['t'])
            risk_list.append(outputs['risk'])
            ci_list.append(outputs['ci_risk'])
        time_list = np.array(time_list)/N
        risks = np.array(risk_list)
        ci = np.array(ci_list)
        ax.plot(time_list, risk_list, lw=2.5, label=name, color=colors[k])
        ax.fill_between(time_list, risks-ci, risks+ci, alpha=0.2, color=colors[k])        
    else:
        continue
    k += 1

ax.set_xlabel(r"Time ($t$)")
ax.set_ylabel(r"Prospective risk")
ax.set_ylim([0, 1])
ax.legend(frameon=False, fontsize=11)
ax.grid(alpha=0.5, ls='--')
ax.set_title('MNIST tasks', fontsize=14)
```
